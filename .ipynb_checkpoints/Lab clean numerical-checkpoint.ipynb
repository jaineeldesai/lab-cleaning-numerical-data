```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
customer.columns= [e.lower().replace(" ",'_') for e in customer.columns]
customer.head()
```

```python
customer["effective_to_date"]=pd.to_datetime(customer["effective_to_date"])
```

```python
customer.head()
```

```python
numerical_df=customer._get_numeric_data()
```

```python
numerical_df
```

```python
numerical_df.nunique()
```

```python
def discrete_continuous(df):
    discrete= [col for col in df.columns if df[col].nunique()<11]
    continuous= [col for col in df.columns if df[col].nunique()>10]
    return df[discrete], df[continuous]

discrete_df, continuous_df= discrete_continuous(numerical_df)
```

```python
corr=numerical_df.corr()
corr
```

```python
sns.heatmap(corr,annot=True)
plt.show()
```

```python
#good correlation between total_claim_amount and monthly_premium_auto
#also good correlation between monthly_premium_auto and customer_lifetime_value
```

```python
def plot_discrete(df):
    for i in df.columns:
        sns.displot(data=df[i],kde=True)
```

```python
plot_discrete(discrete_df)
```

```python
def plot_continuous(df):
    for i in df.columns:
        sns.boxplot(data=df[i])
        plt.ylabel(i)
        plt.show()

plot_continuous(continuous_df)
```
